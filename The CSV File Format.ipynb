```python
filename1 = "sitka_weather_2018_full.csv"

with open(filename1) as f:
    reader1 = csv.reader(f)
    header_row1 = next(reader1)
    
    highs = []
    lows = []
    for column in reader1:
        highs.append(column[8])
        lows.append(column[9])
    print(highs)
    print(lows)
```

```python
# convert string list to integer

highs_int = []
lows_int = []
for i in highs:
    if i == '':
        continue
    high = int(i)
    highs_int.append(high)
for v in lows:
    if v == '':
        continue
    low = int(v)
    lows_int.append(low)
```

```python
print(highs_int)
```

```python
print(lows_int)
```

```python
from matplotlib import pyplot as plt

fig = plt.figure(dpi = 128, figsize=(10,6))
plt.plot(highs_int, c ="red")

plt.title("Daily high temperature, 2018", fontsize=24)
plt.xlabel('',fontsize=16)
plt.ylabel("Temperature (F)", fontsize=16)
plt.tick_params(axis='both', which='major', labelsize=16)

plt.show()
```

```python
from matplotlib import pyplot as plt

fig = plt.figure(dpi = 128, figsize=(10,6))
plt.plot(lows_int, c ="purple")

plt.title("Daily low temperature, 2018", fontsize=24)
plt.xlabel('',fontsize=16)
plt.ylabel("Temperature (F)", fontsize=16)
plt.tick_params(axis='both', which='major', labelsize=16)

plt.show()
```

```python
highs_int_ed = []

for n in highs:
    high = int(n)
    highs_int_ed.append(high)
highs_int_ed
```

```python
# Plotting date
filename2 = "sitka_weather_2018_full.csv"
with open(filename2) as f:
    reader2 = csv.reader(f)
    dates  = []
    for column in reader2:
        if column[2] == 'DATE':
            continue
        column[2] = column[2].replace('/', '-')
        #print(column[2])
        current_date = datetime.strptime(column[2], '%m-%d-%Y')
        dates.append(current_date)
    
fig = plt.figure(dpi=128, figsize=(10,6))
plt.plot(dates, highs_int_ed, c='grey')

plt.title("Daily high temperature with dates, 2018", fontsize=24)
plt.xlabel('',fontsize=16)
plt.ylabel("Temperature (F)", fontsize=16)
fig.autofmt_xdate()
plt.tick_params(axis='both', which='major', labelsize=16)

plt.show()
```
