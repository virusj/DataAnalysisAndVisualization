```python
import pandas as pd
import matplotlib.pyplot as plt
import csv
import seaborn as sns
```

```python
file = "death_valley_2018_full.csv"
with open(file) as f:
    reader = csv.reader(f)
    header_list = next(reader)
    print(header_list)
    for index , columns in enumerate(header_list):
        print(index,columns)
    
```

```python
file1 = "death_valley_2018_full.csv"
with open(file1) as f:
    header1 = csv.reader(f)
    header_row = next(header1)
    
    hight_death = []
    low_death = []
    for column in header1:
        hight_death.append(column[6])
        low_death.append(column[7])
    print(hight_death)
    print(low_death)
    
```

```python
type(hight_death)
```

```python
type(low_death)
```

```python
hight_death_int = []
low_death_int= []
for i in hight_death:
    h = int(i)
    hight_death_int.append(h)
for v in low_death:
    l = int(v)
    low_death_int.append(l)
```

```python
print(hight_death_int)
print(low_death_int)
```

```python
plt.figure(figsize=(10,6))
plt.plot(hight_death_int, c = 'red')
plt.title("Daily Death Valley 2018 ",fontsize=24 , c ="blue")
plt.xlabel("Death Valley", fontsize = 18 ,c = 'red')
plt.ylabel("count" , c='red' , fontsize =18)
plt.tick_params(axis='both', which='major', labelsize=16)
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.plot(low_death_int, c= 'purple')
plt.title("Daily Low Death Valley", fontsize =24 , c = 'green')
plt.xlabel('Low Death Vallay',c = 'red',fontsize = 18)
plt.ylabel('count', c= 'red', fontsize = 18)
plt.show()
```

```python
dates = []
with open(file1) as f:
    head = csv.reader(f)
    dates = []
    
    for j in head:
        if j[2] == 'DATE':
            continue
        j[2]=j[2].replace('/','-')
        
        current_date = datetime.strptime(j[2], '%Y-%m-%d')
        dates.append(current_date)

plt.figure(dpi=128 , figsize =(10,6))
plt.plot(dates,hight_death_int)
plt.plot(dates,low_death_int)
plt.fill_between(dates, hight_death_int, low_death_int, facecolor='blue', alpha=0.1)
plt.tick_params(axis='both', which='major', labelsize=16)
plt.show()
       
```
